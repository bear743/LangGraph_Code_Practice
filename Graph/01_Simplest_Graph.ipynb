```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    message: str
```

```python
def compliment(state: AgentState) -> AgentState:
    state['message'] = state['message'] + ", you're doing an amazing job learning LangGraph!"
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node('compliment_agent', compliment)
graph.set_entry_point('compliment_agent')
graph.set_finish_point('compliment_agent')

app = graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({'message': 'Bob'})
```
